```python
import requests
from bs4 import BeautifulSoup
```

```python
link = "http://sos.ri.gov/openmeetings/?page=view_entity&id=4541"
```

```python
f = requests.get(link)
```

```python
print (f.text)
```

```python
soup = BeautifulSoup(f.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
import string
```

```python
links = []
for link in soup.find_all('a'):
    href = link.get('href')
    if type(href) is str:
        if '/openmeetings/?page=meeting&id=' in href:
            if len(href) == 37:
                links.append(href)
```

```python
for href in links:
    url = 'http://sos.ri.gov' + href
    print(url)
```
